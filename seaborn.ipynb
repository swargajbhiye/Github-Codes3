```python
# introduction to seaborn # seaborn is very powerful visualzation (statistic visualization)
import matplotlib.pyplot as plt   # visualization 
import seaborn as sns             #distribution visualization 

%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4

import warnings
warnings.filterwarnings('ignore')
```

```python
store.head()
```

```python
vis1 = sns.distplot(store["InternetUsers"])  #distributions
```

```python
vis1 = sns.displot(store["InternetUsers"], bins=10)
```

```python
vis2 = sns.boxplot(data = store, x="IncomeGroup", y="BirthRate")
```

```python
vis3 = sns.lmplot(data = store, x = "InternetUsers" , y = "BirthRate" , fit_reg = True) #lm - linear model
```

```python
vis3 = sns.lmplot(data = store, x = "InternetUsers", y = "BirthRate" , fit_reg = False)
```

```python
vis5 = sns.lmplot(data = store, x = "InternetUsers" , y = "BirthRate", fit_reg = False )
```
